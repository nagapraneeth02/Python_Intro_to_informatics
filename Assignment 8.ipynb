```python
#2
import numpy as np
import pandas as pd
df1 = pd.read_csv('voting_data.csv')
cd_A_votes = df1['Candidate A Votes'].values
Total_votes = df1['Total Votes'].values
print(Total_votes)
print(cd_A_votes)
r_sqval = []
for deg in range(1, 11,2):
    coeff = np.polyfit(cd_A_votes, Total_votes, deg)
    predicted_votes = np.polyval(coeff, cd_A_votes)
    print("Predicted votes for Coeff:",deg,";",predicted_votes)
    ss_residual = np.sum((Total_votes - predicted_votes)**2)
    print(ss_residual)
    ss_total = np.sum((Total_votes - np.mean(Total_votes))**2)
    r_sq = 1 - (ss_residual / ss_total)
    r_sqval.append(r_sq)
optimal_degree = 2*np.argmax(r_sqval) + 1
optimal_r_sq = r_sqval[optimal_degree - 1]
print(f"A polynomial of degree {optimal_degree} with an R-squared value of {optimal_r_sq:.4f}.")
```

```python
#3
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def polynomial_eq(x, a, b, c, d):
    return a * x**3 + b * x**2 + c * x + d

x_values = np.linspace(-3, 3, 50)
noise = np.random.normal(0, 1, 50) 
y_values = polynomial_eq(x_values, 4, 5, 5, 8) + noise
popt, pcov = curve_fit(polynomial_eq, x_values, y_values)
y_fit = polynomial_eq(x_values, *popt)
plt.plot(x_values, y_values, 'o', label="data")
plt.plot(x_values, y_fit, label='model')
plt.legend()
plt.show()


```

```python
#4a
import pandas as pd
import matplotlib.pyplot as plt
df2 = pd.read_csv('voting_data.csv')
cd_A_votes = df2['Candidate A Votes']
cd_B_votes = df2['Candidate B Votes']
plt.hist(cd_A_votes, alpha=0.5, label='Candidate A Votes', color='yellow')
plt.hist(cd_B_votes, alpha=0.5, label='Candidate B Votes', color='black')
plt.title('Histogram of Candidate A and Candidate B Votes')
plt.xlabel('No.of Votes')
plt.ylabel('Frequency')
plt.legend()
plt.show()

```

```python
#4b
import pandas as pd
from scipy.stats import ttest_ind
df3 = pd.read_csv('voting_data.csv')
cd_A_votes = df3['Candidate A Votes'].values
cd_B_votes = df3['Candidate B Votes'].values
t_statistic, p_value = ttest_ind(cd_A_votes, cd_B_votes)
if p_value < 0.05:
    print("The means differ with a p-value of", p_value)
else:
    print("The means are not differ with a p-value of", p_value)

```
