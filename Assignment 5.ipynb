1. Write a Python program to generate the following plot with a legend, title, & labels of your choice (3 points)

```python
import matplotlib.pyplot as plt

# Data
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
sales = [0.3, 0.4, 1.7, 2.95, 3.1, 2.05, 1.9, 1.65, 1.8, 1.3, 1.35, 0.5]

# Create plot
plt.plot(months, sales, label="Number Sold")

# Add labels and title
plt.xlabel("Month")
plt.ylabel("Sales (in millions)")
plt.title("Monthly Sales Report")

# Add legend
plt.legend()

# Display plot
plt.show()
```

3. Write a python program that plots the below in one pyplot figure with appropriate axis scales, as shown below with y-axis values = [0,13,8,11,9,17,9,11,25]. (4 points)

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
y = [0, 13, 8, 11, 9, 17, 9, 11, 25]
x = range(len(y))

# Create a figure with subplots
fig, axs = plt.subplots(2, 2, figsize=(8, 6))

# Histogram
axs[0, 0].hist(x, bins=7, color='blue', edgecolor='yellow')
axs[0, 0].set_title('Histogram')
axs[0, 0].set_xlabel('Value')
axs[0, 0].set_ylabel('Frequency')

# Line chart
axs[0, 1].plot(x, color='red')
axs[0, 1].set_title('Line Chart')
axs[0, 1].set_xlabel('Index')
axs[0, 1].set_ylabel('Value')

# Scatter plot
axs[1, 0].scatter(x, y, color='green')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('Index')
axs[1, 0].set_ylabel('Value')

# Bar chart
axs[1, 1].bar(x, y, color='orange')
axs[1, 1].set_title('Bar Chart')
axs[1, 1].set_xlabel('Index')
axs[1, 1].set_ylabel('Value')

# Add spacing between subplots
plt.tight_layout()

# Display the chart
plt.show()
```

Create a pie chart using data given below, with labels: Flavored milk whole, Eggnog and buttermilk, Plain 2% milk, Plain whole milk, Plain 1% milk and Skim milk; and the numbers associated with each label are 9%, 1%, 35%, 27%, 14% and 14% respectively. Give each part a different color (not required to be the same as shown below), a title "U.S. per capita fluid milk availability, gallons, shares, 2013”, list labels, and respective percentage values on the chart. (3 points)

```python
import matplotlib.pyplot as plt

data = [9, 1, 35, 27, 14, 14]
labels = ["Flavored milk whole", "Eggnog and buttermilk", "Plain 2% milk", "Plain whole milk", "Plain 1% milk", "Skim milk"]
colors = ["red", "orange", "yellow", "green", "blue", "purple"]

plt.pie(data, labels=labels, colors=colors, autopct="%1.1f%%", startangle=90)
plt.title("U.S. per capita fluid milk availability,gallons,shares,2013")

plt.show()
```

7. Write a python program that connects to the MySQL database, which has the table imported from the script Revenue.py Create a horizontal bar chart of the top 5 highest revenue generating products in descending order. (4 points)

```python
import MySQLdb
import matplotlib.pyplot as plt

myvars = {}
with open("singpall-mysql-password") as myfile:
    for line in myfile:
        name, var = line.partition(":")[::2]
        myvars[name.strip()] = var.strip()

conn = MySQLdb.connect(host="localhost", user=myvars['DB username'], passwd=myvars['DB password'], db=myvars['DB databasename'])
cursor = conn.cursor()

sql = cursor.execute('SELECT product_name, price FROM Revenue ORDER BY price DESC Limit 5');
results = cursor.fetchall()

product_name = [row[0] for row in reversed(results)]
revenue = [row[1] for row in reversed(results)]

plt.barh(product_name, revenue, color="pink", edgecolor='purple')
plt.title("Top 5 highest revenue-generating products")
plt.xlabel("Revenue")
plt.ylabel("Product Name")
plt.show()
```
